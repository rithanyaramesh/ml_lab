# Scale the data by using MinMaxScaler

```python
cols = X.columns
```

```python
from sklearn.preprocessing import MinMaxScaler

ms = MinMaxScaler()

X = ms.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns=[cols])
```

```python
X.head()
```

#Using Dendogram to find best number of clusters for this data

```python


dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Euclidean distances')
plt.show()

```

Interpreting the dendrogram to decide the number of clusters:

Identify the tallest vertical lines that don't intersect any other clusters horizontally. These lines represent the largest distances that any two clusters are merged into one.

Draw a horizontal line through the tallest unbroken vertical lines. This line indicates a threshold distance beyond which you would not want to cluster data points together.

Count the number of vertical lines that are intersected by the horizontal line. This number represents the optimal number of clusters.

Therefore 2 is the optimal number of clusters.

#Using Hierarchical Clustering Model

```python
# Training the Hierarchical Clustering model on the dataset
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)

# Assign cluster labels to the DataFrame
df['cluster'] = y_hc


```

#Identifying the status_type (Majority) of each cluster

```python
# Group the DataFrame by cluster labels and status_type, count occurrences, and find the majority status_type
majority_status_types = df.groupby(['cluster', 'status_type'])['status_type'].count().groupby(level=0).idxmax().str[1]

# Print the majority status_type of each cluster
print("Majority status_type of each cluster:")
print(majority_status_types)
```

**Inference**:


The majority status_type of cluster 0 is 1 (photo)


The majority status_type of cluster 1 is 1 (video)

#Visualization of Clusters


```python
# Convert DataFrame to NumPy array
X_array = X.values

# Plot clusters using boolean indexing
plt.scatter(X_array[y_hc == 0, 1], X_array[y_hc == 0, 3], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X_array[y_hc == 1, 3], X_array[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.title('Clusters of Facebook Live posts')
plt.legend()
plt.show()
```
